```python
import numpy as np
import pandas as pd

total_time = 150
```

```python
Y = [0]

for t in range(1, total_time):
    Y.append( 5*np.cos(6*t/total_time) + np.random.normal(0,1) )
#     Y.append(Y[t-1] + 30*np.random.normal(0,1))

```

```python
def exp_smooth(Y,alpha):
    smooth = [0]
    for t in range(1, len(Y)):
        smooth.append( (1-alpha)*Y[t] + (alpha)* smooth[t-1])
        
    return smooth
```

```python
def objective(Y,alpha,pen_parameter):
    data = pd.DataFrame({'Y':Y, 'est':exp_smooth(Y,alpha)})
    data['est'] = data['est'].shift(1)
    data = data.dropna()
    
    sum_squares = np.square(data['Y'] - data['est']).sum()
    return sum_squares + pen_parameter*alpha
```

```python
def minimal_alpha(Y,pen_parameter):
    obj_df = pd.DataFrame(columns = ['alpha','value'])
    
    obj_df['alpha'] = np.linspace(0,1,20)
    obj_df['value'] = obj_df['alpha'].apply(lambda s: objective(Y,s,pen_parameter))
    
    min_val = obj_df['value'].min()
    return obj_df.set_index('value').loc[min_val,'alpha']
```

```python
penalty_parameter = 250

obj_df = pd.DataFrame(columns = ['alpha','value'])

obj_df['alpha'] = np.linspace(0,.8,20)
obj_df['value'] = obj_df['alpha'].apply(lambda s: objective(Y,s,penalty_parameter))

obj_df.plot.scatter(x='alpha',y='value')
```

```python
pen_grid = np.linspace(0,500,100)

est_df = pd.DataFrame(columns = ['lambda','alpha_est'])

est_df['lambda'] = pen_grid
est_df['alpha'] = est_df['lambda'].apply(lambda s: minimal_alpha(Y,s))

est_df.plot.scatter(x='lambda',y='alpha')
```
